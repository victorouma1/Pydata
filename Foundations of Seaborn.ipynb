```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.info()
```

```python
sns.set_style('darkgrid')
sns.set_palette('Set1')
plt.figure(figsize = (10,6))
ax = sns.barplot(x='day',y='total_bill', data=tips, errorbar = None, hue = 'sex')

ax.set_title('Average Total by Day and Gender', fontsize = 16, fontweight = 'bold')
ax.set_xlabel('Day of the Week', fontsize = 14)
ax.set_ylabel('Average Total Bill', fontsize = 14)
ax.legend(title = 'Gender', loc = 'upper right', fontsize = 10, title_fontsize = 12)
plt.xticks(rotation = 45)

plt.show()
```
