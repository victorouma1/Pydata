Combining Datasets: Concat and Append

```python
import pandas as pd
import numpy as np
```

```python
def make_df(cols, ind):
    """Quickly make a DataFrame"""
    data = {c: [str(c) + str(i) for i in ind]
            for c in cols}
    return pd.DataFrame(data, ind)

make_df('ABC',range(3))
```

```python
class display(object):
    """Display HTML representation of multiple objects"""
    template = """<div style="float: left; padding: 10px;">
    <p style='font-family:"Courier New", Courier, monospace'>{0}</p>{1}
    </div>"""
    def __init__(self, *args):
        self.args = args

    def _repr_html_(self):
        return '\n'.join(self.template.format(a, eval(a)._repr_html_())
                         for a in self.args)

    def __repr__(self):
        return '\n\n'.join(a + '\n' + repr(eval(a))
                           for a in self.args)
```

```python
series_1 = pd.Series(['A','B','C'], index = [1,2,3])
series_2 = pd.Series(['D','E','F'], index = [1,2,3])

pd.concat([series_1,series_2])
```

Merge and joins

```python
df1 = pd.DataFrame({'employee':['Bob','Lake','Lisa','Sue'],'group':['Accounting','Engineering','Engineering','HR']})
df2 = pd.DataFrame({'employee':['Lisa','Bob','Lake','Sue'],'hire_date':[2004,2008,2012,2014]})
display('df1','df2')
```

```python
df3 = pd.merge(df1,df2)
df3
```

```python
df4 = pd.DataFrame({'group':['Accounting','Engineering','HR'], 'supervisor':['John','Doe','Mike']})
display('df3','df4')
```

```python
pd.merge(df3,df4)
```

```python
df5 = pd.DataFrame({'group':['Accounting','Accounting','Engineering','Engineering','HR','HR'],'skills': ['math','spreadsheets','software','math','spreasheets','communication']})
display('df4','df5','pd.merge(df4,df5)')
```

```python
df7 = pd.DataFrame({'name':['Peter','Paul','Mary'],'food':['fish','beans','bread']},columns = ['name','food'])
df8 = pd.DataFrame({'name':['Mary','Joseph'],'drink':['wine','bear']},columns = ['name','drink'])
display('df7','df8','pd.merge(df7,df8,how = "outer")')
```

```python
df9 = pd.DataFrame({'name':['Bob','Jake','Lisa','Sue'], 'rank':[1,2,3,4]})
df10 = pd.DataFrame({'name':['Bob','Jake','Lisa','Sue'], 'rank':[3,1,4,2]})

display('df9','df10','pd.merge(df9,df10,left_on="name",right_on="name")')
```

```python
pop = pd.read_csv('state-population.csv')
areas = pd.read_csv('state-areas.csv')
abbrevs = pd.read_csv('state-abbrevs.csv')

display('pop.head()','areas.head()','abbrevs.head()')
```

```python
state_and_abbr = pd.merge(pop,abbrevs,how = "outer",left_on = "state/region",right_on = "abbreviation")
all = pd.merge(state_and_abbr,areas,how = "left",on = "state")
all
```

```python
all.isnull().sum()
```

```python
all[['population']]
```

```python
all[all['population'].isnull()].head()
```

```python
all.loc[all['state'].isnull(),'state/region'].unique()
```

```python
all.loc[all['state/region'] == 'PR', 'state'] =  'Puerto Rico'
all.loc[all['state/region'] == 'USA','state']  = 'United States'

all.isnull().sum()
```

```python
all = all.drop(columns = ["abbreviation"])
```

```python
all.isnull().sum()
```

```python
all['state'][all['area (sq. mi)'].isnull()].unique()
```

```python
all.dropna(inplace = True)
all.head()
```
